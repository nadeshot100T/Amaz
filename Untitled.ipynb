```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')


df = getDF(r"C:\\Users\\rajat\\Desktop\\PRO\\NN\\Luxury_Beauty_5.json.gz")
df['original_review_length']  = (df['reviewText'].str.len())
df
```

```python
punc = str.maketrans('', '', string.punctuation)
#apply the empty mapping table to each element of the series where x is the review for one document.
reviews['reviewText'] = reviews['reviewText'].apply(lambda x : ' '.join(word.translate(punc) for word in x.split()))

#removing words that is non alpha
reviews['reviewText'] = reviews['reviewText'].apply(lambda x: ' '.join(word for word in x.split() if word.isalpha()))

#making all words to be lowercase
reviews['reviewText'] = reviews['reviewText'].apply(lambda x: ' '.join(word.lower() for word in x.split()))

#list of stop words
stop = stopwords.words('english')

#removing the stop words
reviews['reviewText'] = reviews['reviewText'].apply(lambda x : ' '.join(word for word in x.split() if word not in stop))

#Lemmatize words to reduce them to their root form. Note: added the pos = 'v' to reduce the incoming word to verb root
lem = WordNetLemmatizer()
reviews['reviewText'] = reviews['reviewText'].apply(lambda x : ' '.join(lem.lemmatize(word, pos = 'v') for word in x.split()))
```

```python
import io
import requests
from PIL import Image
#brightness
def rgb2hsv(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return hsv

def getImageVar(imgPath):
    img2gray = cv2.cvtColor(imgPath, cv2.COLOR_BGR2GRAY)
    imageVar = cv2.Laplacian(img2gray, cv2.CV_64F).var()
    return imageVar

lenR = ival.shape[0]
result = pd.DataFrame(columns=['warm_hue','saturation','brightness','contrast','image_clarity'])
fuck = 0
for i in range(0, lenR):
    url = ival[i] #image choose
    url = url[0]
    #type(url) str
    fuck = i
    r = requests.get(url, stream=True)
    try:
        aux_im = Image.open(io.BytesIO(r.content))
    except Exception as e:
        print(i)
        result.loc[i] = [-1] + [-1] + [-1] + [-1] + [-1]
        continue
    #im = Image.open(requests.get(url, stream=True).raw)
    image = aux_im.resize((80,80))
    image = np.array(image)
    
    if image.ndim !=3:
        print(i)
        result.loc[i] = [-1] + [-1] + [-1] + [-1] + [-1]
        continue
    hsv = rgb2hsv(image)
    h, s, v = hsv[:, :, 0], hsv[:, :, 1], hsv[:, :, 2]
    
    #warm_hue #pixel hues that fall outside the cool range(30–110)
    warm_hue_matrix = np.where((h >30) & (h<110), 0, h)
    nonzero = np.count_nonzero(warm_hue_matrix!=0)
    Tsum = np.sum(warm_hue_matrix)
    warm_hue_value = Tsum/nonzero
    warm_hue_value_res = 0
    if nonzero == 0:
        warm_hue_value_res = 0
    else :   
        warm_hue_value_res = warm_hue_value
    
    #saturation
    #result["saturation"] =  s.mean() 
    
    #brightness
    #result["brightness"] =  v.mean()
    
    #Contrast of Brightness
    #result["contrast"] =  np.std(v) 
    
    #image clarity / Sharpness
    #result["image_clarity"] = getImageVar(image)
    
    result.loc[i] = [warm_hue_value_res] + [s.mean()] + [v.mean()] + [np.std(v)] + [getImageVar(image)]
```
